```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
mpl.rcParams['font.size'] = 11

fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection='3d')


xs = np.arange(1,13)
ys1 = [14,17,11,13,24,9,6,15,13,18,20,16]
ys3 = [0,2,2,0,1,1,0,0,1,2,7,1]
ys2 = [0,3,1,1,2,1,1,1,1,3,3,2]


z1=2  
color =plt.cm.Set2((np.arange(plt.cm.Set2.N)))  
p1=ax.bar(xs, ys1, z1, zdir='y', color='r', alpha=0.7) 

z2=3
color =plt.cm.Set2((np.arange(plt.cm.Set2.N)))  
p1=ax.bar(xs, ys2, z2, zdir='y', color='g', alpha=0.7) 

z3=4
color =plt.cm.Set2((np.arange(plt.cm.Set2.N)))  
p1=ax.bar(xs, ys3, z3, zdir='y', color='b', alpha=0.7) 

ax.xaxis.set_major_locator(mpl.ticker.FixedLocator(xs))
ax.set_xlabel('Month (2013)')
ax.set_ylabel('Group pattern')
ax.set_zlabel('Number of attacking groups')

plt.show()
```

```python
fig = plt.figure(figsize = (15, 15))
ax1 = fig.add_subplot(111)
ax1.xaxis.set_major_locator(mpl.ticker.FixedLocator(xs))
ax1.set_xlabel('Month (2013)', fontsize=15)
ax1.set_ylabel('Number of attacking groups', fontsize=15)
ax1.set_title('Group pattern (Three vandal)', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

p1=ax1.bar(xs, ys2, color='g') 
plt.show()
```
